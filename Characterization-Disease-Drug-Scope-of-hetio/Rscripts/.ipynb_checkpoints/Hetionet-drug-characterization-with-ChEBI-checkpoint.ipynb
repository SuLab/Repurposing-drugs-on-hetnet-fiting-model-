### Creating the ontology graph

```r
fileontology <- read.csv("../out/chebi_ontologyChildParent.tab", sep = "\t", header = F)
edges <- fileontology[c(1,2)]
nodes<- as.data.frame(unique(c(unique(as.character(edges[, 1])), unique(as.character(edges[,2])))))
colnames(nodes) <- "node"
graph.full = graph.data.frame(edges, directed=T, vertices=nodes)
nodes_int = vcount(graph.full) # number of nodes
edges_int = ecount(graph.full) # number of edges
cat('Full graph counts:\n')
cat('Number of nodes:', nodes_int)
cat('\nNumber of edges:', edges_int)

graph <- igraph::simplify(graph.full)
nodes_int = vcount(graph)  
edges_int = ecount(graph)
cat('\n\nSimplified graph counts:')
cat('\nNumber of nodes:', nodes_int)
cat('\nNumber of edges:', edges_int)
```

### Input: parent classes

### ChEBI has three upper classes:
    - Chemical entity
    - Role
    - Subatomic particle
    
Drugs in Hetionet only classify to the upper class 'Chemical Entity' using the obo ontology, which gives chemical structure information. 

The upper class 'Role' has two interesting subclasses: 'Biological role' and 'Chemical role'. This roles classification are not available using the obo ontology, but are available on the ChEBI website.

Ontology: http://bioportal.bioontology.org/ontologies/CHEBI?p=classes&conceptid=root
ChEBI Website: https://www.ebi.ac.uk/chebi/chebiOntology.do?chebiId=CHEBI:7903
Role Branch: http://www.ebi.ac.uk/ols/ontologies/chebi/terms?iri=http%3A%2F%2Fpurl.obolibrary.org%2Fobo%2FCHEBI_59897

http://www.ebi.ac.uk/ols/ontologies/chebi/terms?iri=http%3A%2F%2Fpurl.obolibrary.org%2Fobo%2FCHEBI_39141

#### Chemical Entity SuperClass

```r
fileparent <- read.csv("../out/chebi_ontology1rstBranchParentsChemicalEntity.tab", sep = "\t", header = F)
parent <- fileparent[c(1)]
parent <- parent$V1
cat('\nNumber of parent classes:', length(unique(parent)))
```

### Input: child classes

### Rephetio
It contains 1552 compounds identified by DrugBank IDs.

Data downloaded from: https://neo4j.het.io/browser/ 

Data: https://github.com/dhimmel/learn/blob/5e61aed9789913d5b3525bb83eba80e0fcc060a4/summary/compounds.tsv

```r
filechild <- read.csv("../out/hetionet-chebids-list.tab", sep = "\t", header = F)
child <- filechild[c(1)]
child <- child$V1
cat('\nNumber of child classes:', length(unique(child)))
child <- intersect(child,nodes$node)
cat('\nNumber of child classes in CHEBI:',length(unique(child)))
```

### Output: child -> parent pathways 

```r
resultsPathways  <- data.frame(t(rep(NA,3)))
colnames(resultsPathways ) <- c("child", "parent", "length")
```

### Algorithm

```r
c <- 1  
l <- 0
for (node1 in child ){     
  for (node2 in parent){       
    sp <- get.shortest.paths(graph,from = node1, to = node2, mode = "out")
    
    if (length(sp$vpath[[1]]) > 0){
      l <- length(sp$vpath[[1]])
      resultsPathways[c, 1] <- node1
      resultsPathways[c, 2] <- node2      
      resultsPathways[c, 3] <- l # original row
      c<- c+1
    }
    cat('\nnode1:',node1,' node2:',node2, ' l:', l)  
  }
}
```

### Save results

```r
str(resultsPathways)
write.csv(resultsPathways,"../out/hetionet-chebidsPathwaysChemicalEntity.csv")
```

### Drug class distribution

```r
hetioPathways <- resultsPathways
hetio.freq = table(sort(hetioPathways$parent))
hetio.freq.df = as.data.frame(hetio.freq)
colnames(hetio.freq.df)[1] = 'chebi_id'
hetio.percent = prop.table(hetio.freq)*100
hetio.percent.df <- as.data.frame(hetio.percent)
colnames(hetio.percent.df) = c('chebi_id', 'Percent')
colnames(fileparent) = c('chebi_id','name')
hetio.percent.df.merge = merge(fileparent, hetio.freq.df, by.x = 'chebi_id', all.x = T)
hetio.percent.df.merge$Freq[is.na(hetio.percent.df.merge$Freq)] = 0
hetio.percent.df.merge = merge(hetio.percent.df.merge, hetio.percent.df, by.x = 'chebi_id', all.x = T)
hetio.percent.df.merge$Percent[is.na(hetio.percent.df.merge$Percent)] = 0
hetio.percent.df.merge = hetio.percent.df.merge[order(-hetio.percent.df.merge$Percent),]
hetio.percent.df.merge
```

### Save graphic
